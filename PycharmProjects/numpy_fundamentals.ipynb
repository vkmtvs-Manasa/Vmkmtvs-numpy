# Array Operations

```python
# we can any mathematical operations in numpy
a1 = np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
```

```python
# SCALERR OPEARTIONS
# arthimetic
a1 ** 2
```

```python
#relational
a1>10
```

```python
# VECTOR OPERATONS
# arthemetic operation
a1 + a2
a1/a2
a1 - a2
```

```python
a1 = np.random.random((3,3))
print(a1)
a1 = np.round(a1*10)
print(a1)
```

```python
# max /min /prod/sum
np.max(a1)
np.min(a1)
np.prod(a1)
np.sum(a1)
```

# 0 -> column
# 1 ->row

```python
#np.min(a1,axis=0)
#np.min(a1,axis=1)
#np.max(a1,axis=0)
np.max(a1,axis=1)
```

```python
#mean/ median /std/ var
np.mean(a1,axis=1)
np.median(a1,axis=1)
np.std(a1,axis=1)
np.var(a1,axis =1)
```

```python
# trigo function
np.sin(a1)
```

```python
# dot product
a2 = np.arange(12).reshape(3,4)
a3= np.arange(12,24).reshape(4,3)
np.dot(a2,a3)
```

```python
# log and exponents
np.exp(a1)
```

```python
# round/floor /ceil
np.round(np.random.random((2,3))*100)
np.floor(np.random.random((3,4))*100)
np.ceil(np.random.random((3,4))*100)
```
